# Extract

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_tallest_structures_in_Thailand'
```

```python
df = pd.read_html(url)
len(df)
```

```python
df = df[0]
```

```python
df.head()
```

```python
df.info()
```

# Transform 

```python
#delete column image
df.drop('Image', axis=1, inplace=True)
```

```python
df.head()
```

```python
#clean Rank column
df["Rank"] = df["Rank"].str.replace('\D','',regex=True).astype('int')
```

```python
#clean Height column
df["Height"] = df["Height"].str.replace(r'[^\d.]','',regex=True).astype('float')
```

```python
#clean Height.1 column
df["Height.1"] = df["Height.1"].str.replace('\D','',regex=True).astype('int')
```

```python
#add Province column
df['Province']= df['Location'].str.split('13',expand = True)[0]
df['Province']= df['Province'].str.split('.',expand = True)[0]
```

```python
#clean Notes column
df['Notes'] = df["Notes"].str.split('[',expand = True)[0]
```

```python
df[df["Notes"].isna()].head()
```

```python
df["Notes"] = df["Notes"].fillna(' ')
```

```python
df.info()
```

```python
#set index
df.set_index('Rank', inplace=True)
```

```python
df.rename(columns={'Height':'Height_m','Height.1':'Height_ft'}, inplace=True)
```

```python
df.head()
```

```python
df = df[['Building','Province','Year','Height_m','Height_ft','Floors','Location','Notes']]
```

```python
df.head()
```

```python
query = """CREATE TABLE IF NOT EXISTS tall_buildings_in_thailand (
            id INT PRIMARY KEY AUTO_INCREMENT, 
            rank INT, 
            movie VARCHAR(225), 
            runtime_min INT, 
            types VARCHAR(225),
            gross_million FlOAT, 
            description VARCHAR(225), 
            year INT)"""
```

```python
cursor.execute(query)
```

```python
query = 'INSERT INTO imdb_ranking (rank, movie, runtime_min, types, gross_million, description, year) VALUES (%s,%s,%s,%s,%s,%s,%s)'
```

```python
for index, row in df.iterrows():
        rank = row['rank']
        movie = row['movie']
        runtime_min = row['runtime(min)']
        types = row['types']
        gross_million = row['gross(M)']
        description = row['description']
        year = row['year']
        cursor.execute(query, (rank, movie, runtime_min, types, gross_million, description, year))
```
